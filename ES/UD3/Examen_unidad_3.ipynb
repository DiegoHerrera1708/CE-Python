# EJERCICIO 2: Control de accesos (Conjuntos) (1,5 puntos)

* invitados → lista de personas invitadas

2. Calcula cuántas personas distintas han sido invitados o asistentes.



````python
# Ejercicio 1
invitados = {"jose", "manuel", "pedro", "maria", "ana"}
asistentes = {"manuel", "pedro", "maria", "paco"}

print(f"Personas que estaban invitadas y han asistido: {invitados & asistentes}")
print(f"Personas que asistieron sin estar invitadas {asistentes - invitados}")
print(f"Personas invitadas que no asistieron {invitados - asistentes}")

# Ejercicio 2
invitados = {"jose", "manuel", "pedro", "maria", "ana"}
asistentes = {"manuel", "pedro", "maria", "paco"}
personas = len(invitados | asistentes)

print(f"Personas distintas: {personas}")
````

# EJERCICIO 3: Editor de texto simple (Pila) (2 puntos)

Simula un editor muy básico que permite deshacer acciones.

Debe mostrar el siguiente menú:    

1. Escribir texto
2. Deshacer última acción    
3. Mostrar texto actual
0.  Salir

Cada vez que se escribe algo, se guarda el estado anterior en una pila.

Si se pulsa “Deshacer”, se vuelve al estado anterior.

Si no hay acciones para deshacer, se debe indicar.


````python
texto = []

while True:
    print('''1. Escribir texto
2. Deshacer última acción
3. Mostrar texto actual
4. Salir
    ''')
    resp = int(input("Elige una opcion: "))
    match resp:
        case 1:
            print("Escribe algo: ", end = "")
            texto.append(input())
        
        case 2: 
            if texto == []:
                print("No se puede desahcer porque no hay nada escrito")
            else:
                texto.pop()
                print("Texto deshecho")
               
        case 3:
            print("Texto:")
            for i in texto:
                print(i)
        case 4 : break
            
    print()
````

# EJERCICIO 4: Gestión de pedidos online (Cola) (2 puntos)

Una tienda online gestiona pedidos por orden de llegada. Debe mostrar el siguiente menú:    
1. Nuevo pedido
2. Procesar pedido
3. Ver pedidos pendientes
4. Cancelar todos los pedidos
0. Salir

Al crear un nuevo pedido se pedirá la descripción del pedido.

Al procesar un pedido se debe mostrar la descripción del pedido y sacarlo de la cola.

Los pedidos se atienden en orden FIFO.

Controlar el caso de no haber pedidos.


````python
from collections import deque

pedidos = deque()

while True:
    print('''1. Nuevo pedido
2. Procesar pedido
3. Ver pedidos pendientes
4. Cancelar todos los pedidos
5. Salir
          ''')
    resp = int(input("Elige una opcion: "))
    match resp:
        case 1:
            print("Añade el proximo pedido: ", end ="")
            pedidos.append(input())
        case 2:
            print("Procesando pedidos...")
            if len(pedidos) != 0:
                pedidos.popleft()
                print("Pedido procesado")
            else:
                print("No hay pedidos")
            
        case 3: 
            string = ""
            for p in pedidos:
                string += p + " "
            print(f"Pedidos: {string}")
        case 4:
            pedidos.clear()
            print("Pedidos cancelados")
        case 5: break
    print()
````

# EJERCICIO 5: Inventario de tienda (Diccionario) (1,5 puntos)

Una tienda guarda sus productos en un diccionario con formato:    

```
{nombre_producto: precio}
```

Crea un programa que permita:
1. Añadir productos.
1. Modificar el precio de un producto.
1. Eliminar productos.
1. Mostrar todos los productos con su precio.
1. Mostrar el producto más caro.


````python
diccionario = {}
while True:
    print('''1. Añadir productos.
2. Modificar el precio de un producto.
3. Eliminar productos.
4. Mostrar todos los productos con su precio.
5. Mostrar el producto más caro.
          ''')
    resp = int(input("Elige una opcion: "))
    print()
    match resp:
        case 1:
            producto = input("Nombre del producto: ")
            precio = float(input("Precio: "))
            diccionario[producto] = precio
        case 2:
            producto = input("Dime el nombre del producto: ")
            if producto not in diccionario.keys():
                print("El producton no existe")
            else:
                precio = float(input("Dime el nuevo precio: "))
                diccionario[producto] = precio
                print("Precio actualizado")
        case 3: 
            producto = input("Que producto quieres que eliminar : ")
            if producto not in diccionario.keys():
                print("Producto no encontrado")
            else:
                diccionario.pop(producto)
                print("Producto eliminado")
        case 4:
            for p in diccionario.keys():
                print(f"Producto: {p} \tPrecio: {diccionario[p]}")
        case 5: 
            mas_caro = max(diccionario.values())
            for p in diccionario.keys():
                if diccionario[p] == mas_caro:
                    print(f"Producto mas caro: {p} \tPrecio: {diccionario[p]}")
    print()
````

# EJERCICIO 6: Estadísticas meteorológicas (Estructuras anidadas) (2 puntos)

Se quiere almacenar la temperatura de 3 ciudades durante 7 días.
1. Crea una estructura adecuada para almacenar los datos basada en arrays multidimensionales.
1. Rellena la estructura con datos de prueba.
1. Calcula la media de cada ciudad.
1. Indica el día más caluroso de cada ciudad.
1. Muestra la ciudad con mayor temperatura media.



````python
from random import randint

# Apartado 1 y 2
temperatura = [[randint(0,50) for _ in range(7)]for _ in range(3)]
print(temperatura)

# Apartado 3
media = [round(sum(x)/len(x), 2) for x in temperatura]
for i in range(len(temperatura)):
    print(f"Media de temperatura de la ciudad {i+1}: {media[i]}")

#Apartado 4
for ciudad in range(len(temperatura)):
    mas_calor = 0
    for d in temperatura[ciudad]:
        if d > mas_calor:
            mas_calor = d
        indice = temperatura[ciudad].index(mas_calor)
    indice_ciudad = temperatura.index(temperatura[ciudad])
    print(f"El dia {indice + 1} es el dia con mas calor de la ciudad {indice_ciudad + 1} ")

# Apartado 5
maximo = max(media)
indice = media.index(maximo)

print(f"La ciudad con mayor temperatura media es la {indice + 1} con {media[indice]} y teniendo unas temperaturas semanales de {temperatura[indice]}")
````
