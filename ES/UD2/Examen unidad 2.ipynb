# EJERCICIO 1

Escribe un programa que pida al usuario una cadena de caracteres y usando indexación (slicing) muestre por pantalla:

1. Los caracteres impares de la cadena (ejemplo: "hola" -> "hl")
2. La segunda mitad de la cadena (ejemplo: "adios" -> "os")
3. La cadena invertida (ejemplo: "hola" -> "aloh")

```python
cadena = input("Introduce una cadena: ")

# apartado 1
for i in range(1, len(cadena) + 1):
    if i % 2 != 0:
        print(cadena[i -1], end="")

# print vacio para evitar el end y hacer un salto de linea
print()

# apartado 2
mitad = len(cadena) // 2

if len(cadena) % 2 == 0:
    print(cadena[mitad:])
else:
    print(cadena[mitad + 1:])
    
# apartado 3
print(cadena[::-1])
```

# EJERCICIO 4
Escribe un programa que pida 2 valores numéricos por pantalla (a,b) y muestre:
1. La suma de los dos valores
1. El primer valor elevado al segundo (ab)
1. El mayor de los dos valores
1. La división entera
1. El resto al dividir a entre b

```python
a = int(input("Introduce un numero: "))
b = int(input("Introduce otro numero: "))

print("La suma de los numeros a y b es ", a + b)
print("El primer valor elevado al segundo es ", a ** b)
if a > b:    
    print("El mayor de los dos es ", a)
else:
    print("El mayor de los dos es ", b)
print("La división entera es ", a // b)
print("El resto al dividir a entre b es ", a % b)
```

EJERCICIO 5

Pide dos números por pantalla y calcula:
1. El resultado del or bit a bit
1. El resultado del and bit a bit
1. Muestra si el bit más a la derecha es 0 o 1


```python
a = int(input("Introduce un numero: "))
b = int(input("Introduce otro numero: "))

or_bit_a_bit = a | b
and_bit_a_bit = a & b
print(f"{a:b}")
print(f"{b:b}")
print("El resultado del or bit a bit es", or_bit_a_bit)
print("El resultado del and bit a bit es", and_bit_a_bit)

if a & 1 == 1:
    print("El bit mas a la derecha de", a, "es 1")
else:
    print("El bit mas a la derecha de", a, "es 0")
    
```
