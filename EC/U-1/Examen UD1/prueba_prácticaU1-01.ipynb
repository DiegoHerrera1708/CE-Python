### 7. Análisis de cadenas

**Ejercicio 7.1:** Pide al usuario que introduzca una lista de números separados por comas. Convierte la entrada en una lista de enteros, y luego realiza lo siguiente:

* Encuentra el número mayor y el menor de la lista (puedes usar la función `max` y `min` sobre la lista).
* Calcula la suma total de los números (puedes usar la función `sum` sobre la lista)
* Muestra todos estos valores en formato:
  "El número mayor es: [mayor]"
  "El número menor es: [menor]"
  "La suma total de los números es: [suma]"

````python
'''Ejercicio 7.1'''
lista_numeros = input("introduce una lista de numeros separados por comas: ")
lista_numeros = [int(num) for num in lista_numeros.split(",")]

mayor = max(lista_numeros)
menor = min(lista_numeros)
suma = sum(lista_numeros)
print(f"El numero mas mayor es: {mayor}")
print(f"El numero mas menor es: {menor}")
print(f"La suma total de los numeros es: {suma}")
````

### 9. Slicing y repetición de cadenas

**Ejercicio 9.1**: Solicita al usuario introducir una frase. Luego, utiliza slicing para obtener y mostrar:

   * Los primeros 3 caracteres de la frase.
   * Los últimos 3 caracteres de la frase.
   * Los tres caracteres de la mitad de la frase (si tiene longitud impar, redondea hacia abajo).

   Después, usa el operador de repetición `*` para repetir la frase introducida 3 veces y muestra el resultado.

````python
'''Ejercicio 9.1'''
frase = input("Introduce una frase: ")
print(frase[:3])
print(frase[-3:])

letra_medio = 0
if len(frase) % 2 == 0:
    letra_medio = int(len(frase)/2-2)
else:
    letra_medio = int(len(frase)/2)-1

print(frase[letra_medio:letra_medio+3])
print(frase*3)
````

### 11. Estructuras de control y manejo de errores

**Ejercicio 11.1**: Escribe un programa que pida al usuario una lista de números separados por comas (ejemplo: "4, 7, -3, 9"). Convierte esa lista de cadenas en una lista de números enteros y calcula la suma de los números positivos. Si la lista contiene algún número negativo, lanza una excepción que diga *"Error: contiene números negativos"*.

  **Pasos a seguir**:

   * Utiliza `input()` para obtener la lista de números.
   * Usa `split(",")` para separar la entrada por comas.
   * Convierte las cadenas en enteros utilizando `map()`, una lista definida por intensión o un bucle `for`.
   * Verifica si algún número es negativo y lanza una excepción si es el caso. Puedes emplear la excepción ValueError (y pasarle un mensaje) pero más puntuación si creas tu propia excepción.
   * Si todos los números son positivos, calcula la suma de los números positivos y muestra el resultado. Puedes emplear la función `sum`. Más puntuación si realizas esta operación de forma concisa.

````python
'''Ejercicio 11.1'''
class ValueNegative(Exception):
    def __init__(self, mensaje):
        super().__init__(mensaje)

numeros = input("Introduce unos numeros separados por comas: ")
numeros = [int(num) for num in numeros.split(",")]
for num in numeros:
    if num < 0:
        raise ValueNegative("Error: contiene numeros negativos")

print("La suma de los numeros positivos es", sum(numeros))
````

**Ejercicio 11.2**: Crea un programa que pida al usuario una cadena de texto y luego cuente cuántas vocales (a, e, i, o, u) aparecen en esa cadena, sin diferenciar mayúsculas de minúsculas. Muestra el número de vocales encontradas.

   **Pasos a seguir**:

   * Usa `input()` para obtener una cadena de texto del usuario.
   * Convierte la cadena a minúsculas utilizando `.lower()`.
   * Utiliza un bucle `for` para recorrer cada carácter de la cadena.
   * Si el carácter es una vocal, incrementa un contador.
   * Al final, muestra el número total de vocales.


````python
'''Ejercicio 11.2'''
texto = input("introduce una cadena de texto")
vocales = ["a", "e", "i", "o", "u"]
contador = 0
for letras in texto:
    if letras.lower() in vocales:
        contador += 1
print("El numero total de vocales es: " + str(contador))
````
